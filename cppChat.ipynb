```python
from langchain_community.document_loaders import TextLoader

loadder = TextLoader("C++_Introduction.txt", encoding="utf-8")

loadder
```

```python
text_documents = loadder.load()
text_documents
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
textSplitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=200)
finalDocuments = textSplitter.split_documents(text_documents)
finalDocuments
```

```python
texts = [doc.page_content for doc in finalDocuments]
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import FAISS

db = FAISS.from_documents(finalDocuments, embeddings)
db
```

```python
query = input("puch bhai tera sawal c++ kom leke ")
docs = db.similarity_search(query)
for i in range(len(docs)):
    print(docs[i].page_content)
```
